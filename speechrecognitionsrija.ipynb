```python
from pydub import AudioSegment

# Convert mp3 file to wav (if the uploaded file is in mp3 format)
if uploaded['WhatsApp Ptt 2025-01-20 at 14.55.17.mp3']:  # replace 'your_audio_file.mp3' with your uploaded file's name
    audio = AudioSegment.from_mp3("WhatsApp Ptt 2025-01-20 at 14.55.17.mp3")
    audio.export("converted_audio.wav", format="wav")

```

```python
import speech_recognition as sr

# Initialize recognizer
recognizer = sr.Recognizer()

# Load the audio file
audio_file = sr.AudioFile('converted_audio.wav')  # Replace with the name of your uploaded .wav file

with audio_file as source:
    audio_data = recognizer.record(source)  # Read the entire file into memory

# Recognize speech using Google Web Speech API
try:
    text = recognizer.recognize_google(audio_data)
    print("Recognized text: ", text)
except sr.UnknownValueError:
    print("Google Speech Recognition could not understand the audio.")
except sr.RequestError:
    print("Could not request results from Google Speech Recognition service.")
```
